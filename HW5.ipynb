```python
import scipy.ndimage as ndimage
kernel = ones((20,20))/20
```

```python
single_dot = zeros((101, 101))
single_dot[30, 50] = 1.0
gauss_kernel = ndimage.gaussian_filter(single_dot, 5)
imshow(gauss_kernel)
colorbar()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
x, y = mgrid[0:101, 0:101]
ax.plot_surface(x,y,gauss_kernel)
ax.plot_surface(x+100, y*1.5+100, gauss_kernel*2)
ax.plot_surface(x+300, y*1.5+200, gauss_kernel*2)
fig.add_axes(ax)
```

```python
from scipy.ndimage.filters import convolve as convolve2
from scipy.misc import lena
img = lena()
src_image = img[:,:]
gauss_blur = convolve2(src_image, gauss_kernel)
new_image = src_image - gauss_blur*3

imshow(new_image, cmap=cm.gray, interpolation= 'nearest')
```

```python
import urllib
urllib.urlretrieve("http://www.metisgroup.com/images/sample_note.jpg", "data/rock.jpg")
img2 = imread('data/rock.jpg')
imshow(img2, cmap = cm.gray)
```

```python
#im = scipy.misc.imread('bike.jpg')
img2 = img2.astype('int32')
dx = ndimage.prewitt (img2, 0)  # horizontal derivative
dy = ndimage.prewitt (img2, 1)  # vertical derivative
mag = numpy.hypot(dx, dy)  # magnitude
mag *= 255.0 / numpy.max(mag)  # normalize (Q&D)
imshow(mag, cmap = cm.gray)
```
