```python
img = imread('data/build1.jpg')
img.shape
img = sum(img.astype(float), axis=2)/4.0
img = where(img > 0.7, -1, 1)
imshow(img, cmap='gray')
colorbar()
```

```python
o = imread('data/grayscale2.png')
o = o.astype(float).sum(axis=-1)/3
o = where(o > 0.5, 1, -1)
imshow(o, cmap=cm.gray, interpolation='nearest')
```

```python
#In cross-correlation, the value of an output element is computed as a weighted sum of neighboring elements.
from scipy.signal import correlate2d
cc = correlate2d(img, o)
#print(cc.min(2))

imshow(cc)
colorbar()
gcf().set_figheight(8)

#subplot(120)
y, x = unravel_index(argmax(cc), cc.shape)
#print(x, y)
#argmin(0) <-- minimum for each row
#argmin(0) <-- minimum for rach column

#extracting the minimum and maximum of the ouput correlation array
print(cc[cc != 0].min(), cc[ cc != 0].max())


#print(cc.argmin(0))
#print(np.min(cc.argmin(1), axis=0))
```

```python
imshow(cc, cmap=cm.gray)
colorbar()
gcf().set_figheight(8)
```

```python
#the idea here is that if the the shadow of an image can be used and perceived as depth

from scipy.ndimage.filters import maximum_filter

imshow(maximum_filter(cc, (10,10)))
gcf().set_figheight(8)
```
